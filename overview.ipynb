```python
%matplotlib inline
import numpy as np
import pandas as pd
import money as m
import matplotlib.pyplot as plt
```

Load the data, group by place with a total for that place.

```python

df = pd.read_csv("results.csv")
grouped = df.groupby("Source").sum()
grouped.reset_index()
grouped = grouped.sort_values('Value', ascending=False)
```

Show in a data table

```python
grouped
```

```python
bar_chart = grouped.plot.bar()
plt.show()
```

```python
df = []
for index, row in grouped.iterrows():
    d = {'source': index, 'value': m.convert_pennies(row.item())}
    df.append(d)
```

```python
df = pd.DataFrame(df)
df
```
